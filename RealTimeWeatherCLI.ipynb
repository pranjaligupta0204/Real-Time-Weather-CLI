```python
import requests

API_KEY = "9820df9253fc4af2bfe171922252011"   # <<-- put your key here
BASE_URL = "https://api.weatherapi.com/v1/current.json"
```

```python
def get_weather(city):
    try:
        params = {
            "key": API_KEY,
            "q": city,
            "aqi": "no"
        }

        response = requests.get(BASE_URL, params=params)
        data = response.json()

        # WeatherAPI error format
        if "error" in data:
            print("City not found or invalid request.")
            return

        current = data["current"]
        location = data["location"]

        print("\n------ Weather Info ------")
        print("City:", location["name"])
        print("Region:", location["region"])
        print("Country:", location["country"])
        print("Temperature:", current["temp_c"], "°C")
        print("Feels Like:", current["feelslike_c"], "°C")
        print("Humidity:", current["humidity"], "%")
        print("Condition:", current["condition"]["text"])
        print("Wind Speed:", current["wind_kph"], "km/h")
        print("--------------------------\n")

    except Exception as e:
        print("Error:", e)
```

```python
while True:
    city = input("Enter city name (or type 'exit'): ").strip()
    if city.lower() == "exit":
        break
    get_weather(city)
```
